```python
## Cria o arquivo users.csv com dados fictícios
import pandas as pd

dados = {
    "nome": ["Alice Souza", "Bruno Lima", "Carla Mendes"],
    "conta": ["1234-5", "5678-9", "9012-3"],
    "cartao": ["Ouro", "Prata", "Platina"],
    "saldo": [1500.00, 320.50, 8700.00],
}

df_entrada = pd.DataFrame(dados)
df_entrada.to_csv("users.csv", index=False)
print("✅ users.csv criado com sucesso!")
df_entrada.head()
```

```python
import pandas as pd

# E — Extração: lê o CSV de origem
df = pd.read_csv("users.csv")

print(f"📥 {len(df)} usuários carregados")
print(f"Colunas: {list(df.columns)}\n")
print(df.to_string(index=False))
```

```python
import anthropic

# Configure sua chave da API Anthropic
client = anthropic.Anthropic(api_key="sua_chave_aqui")

def gerar_mensagem(nome, cartao, saldo):
    """T — Transformação: usa IA para gerar mensagem personalizada."""
    prompt = (f"Você é um especialista em marketing financeiro do Santander. "
              f"Crie 1 mensagem curta (máx 2 frases) e motivadora sobre "
              f"gestão financeira para o(a) cliente {nome}, que possui "
              f"cartão {cartao} e saldo de R$ {saldo:.2f}. "
              f"Seja personalizado, positivo e use o nome do cliente.")

    resposta = client.messages.create(
        model="claude-opus-4-5",
        max_tokens=150,
        messages=[{"role": "user", "content": prompt}]
    )
    return resposta.content[0].text.strip()

# Aplica a transformação em cada linha do DataFrame
df["mensagem_ia"] = df.apply(
    lambda row: gerar_mensagem(row["nome"], row["cartao"], row["saldo"]),
    axis=1
)

print("✅ Mensagens geradas!\n")
for _, row in df.iterrows():
    print(f"👤 {row['nome']} ({row['cartao']})")
    print(f"   {row['mensagem_ia']}\n")
```

```python
# L — Carregamento: persiste o resultado enriquecido
df.to_csv("users_com_mensagens.csv", index=False, encoding="utf-8-sig")

print("💾 Arquivo salvo: users_com_mensagens.csv")
print(f"✅ Pipeline ETL concluído: {len(df)} registros processados.")
df[["nome", "cartao", "mensagem_ia"]].head()
```
